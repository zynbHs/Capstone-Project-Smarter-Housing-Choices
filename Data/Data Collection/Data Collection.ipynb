```python
import requests
from bs4 import BeautifulSoup
import csv
import re
```

```python
properties_detail = []
property_id = 0
```

```python
def scrape(link):
    # parse the page
    page = requests.get(f"{link}")
    src = page.content
    soup = BeautifulSoup(src, "lxml")

    # all property cards on the page
    properties_link = soup.find_all("article", {'class': 'styles-module_property-card__wrapper__OkXlW'})


    for i in range(len(properties_link)):
        page_link = properties_link[i].find('a')['href']
        src1 = requests.get(f"{page_link}").content
        soup1 = BeautifulSoup(src1, "lxml")

        # unique id for each property
        global property_id
        property_id += 1

        # get offer type (buy or rent)
        offer = soup.find('div', {'class': 'styles_filters-form__tag-container__E1Ctb'}).find('span').text

        # get property type: apartment, villa, ...
        ptype = soup1.find('p', {'data-testid': 'property-details-type'})
        property_type = ptype.get_text(strip=True) if ptype else ""

        # get price
        price = soup1.find('p', {'data-testid': 'property-price'})
        property_price = price.get_text(strip=True) if price else ""


        # get if price includes water and electricity
        include_w_e = properties_link[i].find('div', {'class': 'tag-module_tag__jFU3w'})
        include_w_e = include_w_e.get_text(strip=True) if include_w_e else ""

        # get property size
        size_tag = soup1.find('p', {'data-testid': 'property-details-size'})
        property_size = size_tag.get_text(strip=True) if size_tag else ""

        # get num of bedrooms
        num_of_bedrooms = soup1.find('p', {'data-testid': 'property-details-bedrooms'})
        num_of_bedrooms = num_of_bedrooms.get_text(strip=True) if num_of_bedrooms else "0"

        # get num of bathrooms
        num_of_bathrooms = soup1.find('p', {'data-testid': 'property-details-bathrooms'})
        num_of_bathrooms = num_of_bathrooms.get_text(strip=True) if num_of_bathrooms else "0"

        # get availability-date
        availability_date = soup1.find('p', {'data-testid': 'property-details-rental-availability-date'})
        availability_date = availability_date.get_text(strip=True) if availability_date else ""

        # get title
        title = soup1.find('h1', {'class': 'styles_desktop_title__j0uNx'})
        title = title.get_text(strip=True) if title else ""

        # get location (area & governorate)
        # get location (area & governorate)
        loc_tag = soup1.find('p', {'class': 'styles-module_map__title__M2mBC'})
        if loc_tag:
          parts = [p.strip() for p in loc_tag.get_text(strip=True).split(",") if p.strip()]
          if len(parts) >= 2:
            area = ", ".join(parts[:-1])   # everything except last
            governorate = parts[-1]        # last part
          else:
            area = parts[0] if parts else ""
            governorate = ""
        else:
          area = ""
          governorate = ""

        # get agent name
        agent_name = soup1.find('p', {'data-testid': 'property-detail-agent-name'})
        agent_name = agent_name.get_text(strip=True) if agent_name else ""


        # get agency
        agency = soup1.find('p', {'data-testid': 'property-detail-broker-name'})
        agency = agency.get_text(strip=True) if agency else ""


        # amenities
        number_amenities = ""
        ame = soup1.find('section', {'data-testid': 'amenities-section'})
        if ame:
            button= ame.find('button', {'data-testid': 'button--secondary'})
            if button:
                text = button.get_text(strip=True)
                match = re.search(r"\((\d+)\)", text)
                if match:
                  number_amenities = int(match.group(1))

            else:
              amenities = ame.find_all('div', {'class':'styles_amenity__c2P5u'})
              number_amenities = len(amenities)




        properties_detail.append({
            'Property_id': property_id,
            'Offer': offer,
            'URL': page_link,
            'Property_type': property_type,
            'Price': property_price,
            'Include_w_e': include_w_e,
            'Title': title,
            'Area': area,
            'Governorate': governorate,
            'Beds': num_of_bedrooms,
            'Baths': num_of_bathrooms,
            'Size': property_size,
            'Availability_date': availability_date,
            'Agent_name': agent_name,
            'Agency': agency
            ,'Amenities': number_amenities
        })


    return properties_detail
```

```python
for i in range(1,562):
    link = f"https://www.propertyfinder.bh/en/rent/properties-for-rent.html?page={i}"
    data = scrape(link)
    print(f"rent: {i}")
print(f"\nCollected {len(data)} properties.")
```

```python
print(len(data))
# save data into csv file
keys = properties_detail[0].keys()
with open('rent.csv', 'w', newline='') as output_file:
    dict_writer = csv.DictWriter(output_file, keys)
    dict_writer.writeheader()
    dict_writer.writerows(properties_detail)
```
